```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('advertising.csv')
print(df.head(5))
print(df.info())

sns.set_style('whitegrid')
```

```python
sns.pairplot(df)
plt.show()
```

```python
X = df['TV'].values
print(type(X))
print(X)
```

```python
X = X.reshape(-1 , 1)
print(type(X))
print(X)
```

```python
Y = df['Sales'].values
print(type(Y))
print(Y)
```

```python
Y = Y.reshape(-1,1)
print(type(Y))
print(Y)
```

```python
np.random.seed(42)

indices = np.random.permutation(len(X))
print(indices)
```

```python
train_indices = indices[40:]
test_indices = indices[:40]

print(len(train_indices), len(test_indices))
```

```python
X_train, X_test = X[train_indices], X[test_indices]
Y_train, Y_test = Y[train_indices], Y[test_indices]
```

```python
ones = np.ones((len(X_train), 1))
print(ones.shape)
```

```python
X_b = np.c_[ones, X_train]
print(X_b.shape)

print(X_b)
```

```python
term_01 = np.linalg.inv(X_b.T @ X_b)
print(term_01)

term_02 = X_b.T @ Y_train

model_weights = term_01 @ term_02
print(model_weights)    
print
```

```python
Y_train_pred = X_b @ model_weights
print(f"Y_train_pred: {Y_train_pred}")
print(f"Y_train_pred shape: {Y_train_pred.shape}")
```

```python
ones = np.ones((len(X_test), 1))
X_test_b = np.c_[ones, X_test]
Y_test_pred = X_test_b @ model_weights
print(f"Y_test_pred: {Y_test_pred}")
print(f"Y_test_pred shape: {Y_test_pred.shape}")
```

```python
def rmse(y, y_hat):
    return np.sqrt(np.mean((y - y_hat) ** 2))
print(f"Train RMSE: {rmse(Y_train, Y_train_pred)}")
print(f"Test RMSE: {rmse(Y_test, Y_test_pred)}")
```

```python
residuals = Y_test - Y_test_pred
print(f"Residuals : {residuals}")
plt.figure(figsize=(6, 5))
plt.scatter(Y_test_pred, residuals)
plt.axhline(0, linestyle='--')  # Add a horizontal line at y=0 for reference
plt.xlabel("Predicted sales")
plt.ylabel("Residuals")
plt.title("Residual vs Predicted values")
plt.show()
```
