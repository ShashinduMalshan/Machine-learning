# **Matrix–Vector Multiplication**

## **1. Concept**

Matrix–Vector Multiplication means multiplying a **matrix** with a **vector**.

genui{"math_block_widget_common_keywords": {"content": "y = A x"}}

* ( A ) → Matrix
* ( x ) → Vector
* ( y ) → Result vector

---

## **2. Example**

### Define Matrix and Vector

```python
import numpy as np

A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

x = np.array([1, 2, 3])
```

---

## **3. Perform Multiplication**

```python
result = np.dot(A, x)
print("Result:", result)
```

---

## **6. Shape Rule (Very Important)**

* Matrix shape → `(m, n)`
* Vector shape → `(n,)`
* Result shape → `(m,)`

👉 Inner dimensions must match!

---

## **7. Simple Explanation**

* Each row of the matrix is multiplied with the vector
* Then summed → gives one value
* So each row produces one result

## **8. Real-World Use**

* Image processing
* Neural networks
* Graphics transformations
* Machine learning models

---


````python
import numpy as np

M = np.array([[4, 3, 1], [1, 2, 5], [6, 7, 4]])
x = np.array([6, 4, 3])

result = M @ x
print("Matrix M:\n", M)
print("Vector x:\n", x)
print("Result:\n", result)
````

````python
from matplotlib import pyplot as plt


original_img = np.random.randint(0, 150, (5, 5))
print("Original Image:\n", original_img)

flatten_img= original_img.flatten()
print("Flattened Image:\n", flatten_img)

identity_matrix = np.eye(25)
print("Identity Matrix:\n", identity_matrix)

M = 2.5 * identity_matrix

print("Matrix M:\n", M)

print("Transformed Image")
transformed_img = M @ flatten_img

reshaped_img = transformed_img.reshape(5, 5)
print("Reshaped Image:\n", reshaped_img)

reshaped_img = np.clip(reshaped_img, 0, 255)

plt.subplot(1,2,1)
plt.imshow(original_img, cmap='gray')
plt.axis('off')
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(reshaped_img, cmap='gray')
plt.axis('off')
plt.title("Transformed Image")
plt.show()



````
