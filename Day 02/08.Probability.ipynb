```python
# sigmoid function

import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

print(sigmoid(0))  # Should return 0.5
print(sigmoid(1))  # Should return approximately 0.731058578630004
print(sigmoid(99999999999999999)) # Should return approximately 0.2689414213699951

arr = np.array([-2, -1, 0, 1, 2])
sigmoid_arr = sigmoid(arr)
print("Sigmoid of array:", sigmoid_arr)

```

```python
desicions = sigmoid_arr > 0.65
print("Decisions based on sigmoid values:", desicions)
```
