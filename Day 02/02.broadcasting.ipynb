```python
import numpy as np
from matplotlib import pyplot as plt

original_img = np.random.randint(0, 255, (5,5))

print (f"Original image array:\n{original_img}")

plt.subplot(1, 2, 1)
plt.imshow(original_img, cmap='gray' ,vmin=0, vmax=255)
plt.title("Original Grayscale Image")
plt.axis('off')


plt.subplot(1, 2, 2)
brightened_img = original_img + np.array([0, 50, 100, 150, 200])
brightened_img = np.clip(brightened_img, 0, 255)
plt.imshow(brightened_img, cmap='gray' ,vmin=0, vmax=255)
plt.title("Brightened Grayscale Image")
plt.axis('off')
```
