```python
import numpy as np
from matplotlib import pyplot as plt

img_1 = np.random.randint(0, 200, (5,5))
img_2 = np.random.randint(0, 200, (5,5))

print (f"Image 1 array:\n{img_1}")
print (f"Image 2 array:\n{img_2}")

plt.subplot(1, 3, 1)
plt.imshow(img_1, cmap='gray' ,vmin=0, vmax=255)
plt.title("Image 1")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_2, cmap='gray' ,vmin=0, vmax=255)
plt.title("Image 2")
plt.axis('off')

img_3 = img_1 + img_2
img_3 = np.clip(img_3, 0, 255)
print (f"Summed image array:\n{img_3}")
plt.subplot(1, 3, 3)
plt.imshow(img_3, cmap='gray' ,vmin=0, vmax=255)
plt.title("Summed Image")
plt.axis('off')
```
