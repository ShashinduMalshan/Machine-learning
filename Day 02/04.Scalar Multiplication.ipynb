```python
import numpy as np
from matplotlib import pyplot as plt

img_1 = np.random.randint(0, 200, (5,5))

print (f"Image 1 array:\n{img_1}")

scalar_multiplication = img_1 * 2.5
scalar_multiplication = np.clip(scalar_multiplication, 0, 255)
print (f"Image 1 after scalar multiplication by 2.5:\n{scalar_multiplication}")

plt.subplot(1, 2, 1)
plt.imshow(img_1, cmap='gray' ,vmin=0, vmax=255)
plt.title("Image 1")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(scalar_multiplication, cmap='gray' ,vmin=0, vmax=255)
plt.title("Image 1 after scalar multiplication")    
plt.axis('off')
```
