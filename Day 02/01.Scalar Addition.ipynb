```python
import numpy as np
from matplotlib import pyplot as plt

original_img = np.random.randint(0, 100, (5,5))

print (f"Original image array:\n{original_img}")

brightened_img = original_img + 150
brightened_img = np.clip(brightened_img, 0, 255)

print (f"Brightened image array:\n{brightened_img}")

plt.subplot(1,2,1)
plt.imshow(original_img, cmap='gray' ,vmin=0, vmax=255)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(brightened_img, cmap='gray' ,vmin=0, vmax=255)
plt.title("Brightened Image")
plt.axis('off')
```
