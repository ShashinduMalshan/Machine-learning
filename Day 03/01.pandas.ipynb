# **Creating a DataFrame using Pandas**

## **1. Import Pandas Library**

```python id="a1b2c3"
import pandas as pd
```

* `pandas` is a powerful library for **data analysis and manipulation**.
* `pd` is just a short alias.

---

## **2. Create a Dictionary**

```python id="d4e5f6"
student = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [20, 21, 22],
    'grade': ['A', 'B', 'C']
}
```

### Explanation

* This is a **Python dictionary**.
* Each key represents a **column name**.
* Each list represents the **data in that column**.

| name    | age | grade |
| ------- | --- | ----- |
| Alice   | 20  | A     |
| Bob     | 21  | B     |
| Charlie | 22  | C     |

---

## **3. Create DataFrame**

```python id="g7h8i9"
df = pd.DataFrame(student)
```

* Converts the dictionary into a **DataFrame (table format)**.
* DataFrame = rows + columns (like Excel table).

---

## **4. Display Data**

```python id="j1k2l3"
print(df)
```

### Output

```id="m4n5o6"
      name  age grade
0    Alice   20     A
1      Bob   21     B
2  Charlie   22     C
```

### ✅ describe

`df.describe()` will give something like:

|       | Age  | Salary  |
| ----- | ---- | ------- |
| count | 3    | 3       |
| mean  | 30.0 | 60000.0 |
| std   | 5.0  | 10000.0 |
| min   | 25.0 | 50000.0 |
| 25%   | 27.5 | 55000.0 |
| 50%   | 30.0 | 60000.0 |
| 75%   | 32.5 | 65000.0 |
| max   | 35.0 | 70000.0 |

 **count** → number of non-null values
 **mean** → average value
 **std** → standard deviation (spread of data)
 **min / max** → smallest / largest values
 **25%, 50%, 75%** → quartiles (25% = Q1, 50% = median, 75% = Q3)

---

### ⚠️ Note:

* `df.describe()` **only summarizes numeric columns by default**.
* To include all columns (including strings), use:

```python
df.describe(include="all")
```

This will give counts, unique values, top value, frequency, etc., for object columns.


````python
print(df.describe())
````

### isnull:

1. `df.isnull()` → returns a **boolean DataFrame** with `True` where values are missing (`NaN`) and `False` otherwise:

```
    Name    Age  Salary
0  False  False   False
1  False   True   False
2   True  False    True
```

2. `.sum()` → adds up `True` values for each column, giving the **number of missing values per column**.

---

### ⚡ Tip:

* If **all columns are filled** (no missing data), this will show `0` for all columns.
* You can use this to **decide how to clean your data**, e.g., fill missing values or drop rows.

---

````python
print(df.isnull().sum())
print(df.info())
````


### 1️⃣ `df.dropna(inplace=True)`

* This **removes all rows that have any missing values (`NaN`)** in your DataFrame.
* `inplace=True` means it **modifies the original `df`** instead of creating a new DataFrame.

````python
df.dropna()
print(df.isnull().sum())
print(df.info())
````
