# Grayscale Image Visualization with NumPy and Matplotlib

In this example, we create a **grayscale image** from a 2D NumPy array and display it using `matplotlib`.

### Step 1: Define the Image Matrix
We define a **5×5 matrix** where each value represents a **pixel intensity** (0 = black, 255 = white):

```python
import numpy as np

img = np.array([
    [10, 247, 197, 55, 23],
    [255, 34, 5, 71, 99],
    [14, 2, 3, 4, 5],
    [255, 20, 39, 44, 155],
    [155, 14, 200, 66, 245]
])

print(f"Image array:\n{img}")
````

**Output Matrix:**

```
[[ 10 247 197  55  23]
 [255  34   5  71  99]
 [ 14   2   3   4   5]
 [255  20  39  44 155]
 [155  14 200  66 245]]
```

Each number in the matrix corresponds to a pixel intensity in the grayscale image.

---

### Step 2: Display the Image

We can visualize the matrix as an image using `matplotlib.pyplot.imshow()`. We use `cmap='gray'` to show it in grayscale.

```python
import matplotlib.pyplot as plt

plt.imshow(img, cmap='gray')  # Display as grayscale
plt.title("Grayscale Image")
plt.axis('off')  # Hide axes for better visualization
plt.show()
```

---

### Explanation:

* **`img`** → 2D NumPy array representing the pixel intensities.
* **`plt.imshow(img, cmap='gray')`** → Displays the array as a grayscale image.
* **`plt.axis('off')`** → Hides the x and y axes for a cleaner image view.
* **Pixel intensity**:

  * 0 → Black
  * 255 → White
  * Values in between → Shades of gray

The resulting image is a **visual representation of the matrix**, where brighter numbers appear lighter and smaller numbers appear darker.


`````python
import numpy as np
import matplotlib.pyplot as plt

img = np.array([[10,247,197,55,23],
                [255,34,5,71,99],
                [14,2,3,4,5],
                [255,20,39,44,155],
                [155,14,200,66,245]])

print (f"Image array:\n{img}")
                
plt.imshow(img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')  # Hide axes
plt.show()
`````
