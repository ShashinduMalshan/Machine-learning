```python
import numpy as np

x_1 = np.array([2000,2,1])

print(f"x_1 feature vector: {x_1}")
print(f"x_1 type: {type(x_1)}")
```

```python
x_2 = np.array([2560,3,1])

sum_x_1_x_2 = x_1 + x_2
print(f"Sum of x_1 and x_2: {sum_x_1_x_2}")

multification = x_1 * x_2
print(f"Product of x_1 and x_2: {multification}")

```

```python
x = np.array([[2000,2,1]
              ,[2560,3,1]
              ,[3000,4,1],
              [3000,4,1],
              [3000,4,1]])

print(f"Feature matrix x:\n{x}")
print (f"Feature matrix x shape: {x.shape}")
```

```python
print(x[0,1]) # Accessing the second element of the first row
print(x[1,0]) # Accessing the first element of the second row
```

```python
print(x[1:4,:]) # Accessing the first two elements of the third row
```

```python
print(x[:3,:2])
```
