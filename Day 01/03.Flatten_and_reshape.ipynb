## **1. Importing Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
```

* `numpy` (np) → used for numerical operations and creating arrays.
* `matplotlib.pyplot` (plt) → used for plotting images and graphs.

---

## **2. Creating a Random Grayscale Image**

```python
img = np.random.randint(0, 256, (5,5))
```

* `np.random.randint(0, 256, (5,5))` generates a **5x5 array** of random integers from 0 to 255.
* In image terms:

  * Each integer represents a **pixel intensity**.
  * `0` → black, `255` → white, values in between → shades of gray.
* So `img` is a **grayscale image** of size 5×5.

```python
print(f"Image array:\n{img}")
print(f"Image shape: {img.shape}")
```

* `.shape` shows the dimensions of the image → `(5,5)` (5 rows, 5 columns).

---

## **3. Displaying the Image**

```python
plt.imshow(img, cmap='gray')
plt.title("Random Grayscale Image")
plt.axis('off')  # Hide axes
plt.show()
```

* `plt.imshow(img, cmap='gray')` → displays the 2D array as a grayscale image.
* `plt.title()` → adds a title to the image.
* `plt.axis('off')` → hides the x and y axes for a cleaner view.
* `plt.show()` → renders the image in the notebook.

This will show a 5x5 random grayscale image.

---

## **4. Flattening the Image**

```python
flattened_img = img.flatten()
```

* `.flatten()` converts a **2D array** into a **1D array** (single row).
* Useful for feeding image data into **machine learning models**.

```python
print(f"Flattened image array:\n{flattened_img}")
print(f"Flattened image shape: {flattened_img.shape}")
```

* Now the shape becomes `(25,)` because (5 \times 5 = 25).
* Each pixel value is preserved but listed in a **single dimension**.

---

### **Visual Summary for Notebook**

| Step | Operation                | Result                |
| ---- | ------------------------ | --------------------- |
| 1    | Create 5x5 random array  | 2D grayscale image    |
| 2    | Show image with `imshow` | Visual 5x5 image      |
| 3    | Flatten image            | 1D array of 25 pixels |

---

````python
import numpy as np
import matplotlib.pyplot as plt

img = np.random.randint(0, 256, (5,5),)

print (f"Image array:\n{img}")
print (f"Image shape: {img.shape}")

plt.imshow(img, cmap='gray')
plt.title("Random Grayscale Image")
plt.axis('off')  # Hide axes
plt.show()

flattened_img = img.flatten()
print (f"Flattened image array:\n{flattened_img}")
print (f"Flattened image shape: {flattened_img.shape}")
````

## **5. Reshaping the Image Back**

```python
reshaped_img = flattened_img.reshape(img.shape)
```

* `.reshape(img.shape)` changes the **1D array** back into a **2D array** with the same shape as `img` (`5×5`).
* Basically, it reverses the flattening process.

```python
print(f"Reshaped image array:\n{reshaped_img}")
print(f"Reshaped image shape: {reshaped_img.shape}")
```

* `reshaped_img` now looks exactly like the original `img` array.
* `.shape` confirms it → `(5, 5)`.

So the flow is:

1. Original 2D image → `(5,5)`
2. Flatten → 1D array `(25,)`
3. Reshape → back to 2D `(5,5)`
