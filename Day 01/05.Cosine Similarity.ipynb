```python
import numpy as np
movie_A = np.array([2.5,0.5])
movie_B = np.array([10.5,5])
user = np.array([5,1])

cosine_similarity_A_and_user = np.dot(movie_A, user) / (np.linalg.norm(movie_A) * np.linalg.norm(user))
print(f"Cosine similarity between Movie A and User: {cosine_similarity_A_and_user}")

cosine_similarity_B_and_user = np.dot(movie_B, user) / (np.linalg.norm(movie_B) * np.linalg.norm(user))
print(f"Cosine similarity between Movie B and User: {cosine_similarity_B_and_user}")
```
