# Grayscale Image Visualization with NumPy and Matplotlib

In this example, we create a **grayscale image** from a 2D NumPy array and display it using `matplotlib`.

### Step 1: Define the Image Matrix
We define a **5×5 matrix** where each value represents a **pixel intensity** (0 = black, 255 = white):

```python
import numpy as np

img = np.array([
    [10, 247, 197, 55, 23],
    [255, 34, 5, 71, 99],
    [14, 2, 3, 4, 5],
    [255, 20, 39, 44, 155],
    [155, 14, 200, 66, 245]
])

print(f"Image array:\n{img}")
````

**Output Matrix:**

```
[[ 10 247 197  55  23]
 [255  34   5  71  99]
 [ 14   2   3   4   5]
 [255  20  39  44 155]
 [155  14 200  66 245]]
```

Each number in the matrix corresponds to a pixel intensity in the grayscale image.

---

### Step 2: Display the Image

We can visualize the matrix as an image using `matplotlib.pyplot.imshow()`. We use `cmap='gray'` to show it in grayscale.

```python
import matplotlib.pyplot as plt

plt.imshow(img, cmap='gray')  # Display as grayscale
plt.title("Grayscale Image")
plt.axis('off')  # Hide axes for better visualization
plt.show()
```

---

### Explanation:

* **`img`** → 2D NumPy array representing the pixel intensities.
* **`plt.imshow(img, cmap='gray')`** → Displays the array as a grayscale image.
* **`plt.axis('off')`** → Hides the x and y axes for a cleaner image view.
* **Pixel intensity**:

  * 0 → Black
  * 255 → White
  * Values in between → Shades of gray

The resulting image is a **visual representation of the matrix**, where brighter numbers appear lighter and smaller numbers appear darker.


`````python
import numpy as np
import matplotlib.pyplot as plt

img = np.array([[10,247,197,55,23],
                [255,34,5,71,99],
                [14,2,3,4,5],
                [255,20,39,44,155],
                [155,14,200,66,245]])

print (f"Image array:\n{img}")
                
plt.imshow(img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')  # Hide axes
plt.show()
`````

# RGB Image Visualization with NumPy and Matplotlib

In this example, we create a **small RGB image** using a 3D NumPy array and display it with `matplotlib`.

---

### Step 1: Define the Image Matrix

Here, we define a **3×2 image**, where each pixel has 3 values representing **Red, Green, Blue (RGB)**:

```python
import numpy as np

img = np.array([
    [[255, 0, 0], [0, 255, 8]],    # Row 0: Red, Green-ish
    [[0, 255, 0], [255, 0, 0]],    # Row 1: Green, Red
    [[0, 0, 255], [255, 255, 0]]   # Row 2: Blue, Yellow
])

print(f"Image array:\n{img}")
````

**Explanation of the array:**

* Each element in the 3D array represents a **pixel**: `[R, G, B]`
* Shape of the array: `(3, 2, 3)`

  * 3 → Number of rows
  * 2 → Number of columns
  * 3 → RGB channels
* Example:

  * `[255, 0, 0]` → Bright **red** pixel
  * `[0, 255, 0]` → Bright **green** pixel
  * `[0, 0, 255]` → Bright **blue** pixel
  * `[255, 255, 0]` → **Yellow** pixel (red + green)

---

### Step 2: Display the RGB Image

```python
import matplotlib.pyplot as plt

plt.imshow(img)          # Display the RGB image
plt.title("RGB Image")   
plt.axis('off')          # Hide axes for clean visualization
plt.show()
```

---

### What Happens

* `plt.imshow()` automatically interprets a 3D NumPy array as an **RGB image**.
* Each pixel’s color is determined by the `[R, G, B]` values (0–255).
* The result is a **tiny 3×2 colored image**:

  * Top row → Red, green-ish
  * Middle row → Green, red
  * Bottom row → Blue, yellow

---
**Note:**
For larger images, you can use arrays with more rows and columns. Each pixel still follows the `[R, G, B]` format.

---


`````python
import matplotlib.pyplot as plt
import numpy as np
img = np.array([[[255,0,0],[0,255,8]],
                [[0,255,0],[255,0,0]],
                [[0,0,255],[255,255,0]]])

print (f"Image array:\n{img}")
plt.imshow(img)
plt.title("RGB Image")
plt.axis('off')  # Hide axes
plt.show()
`````

# Random Grayscale Image Visualization

We can generate a **random grayscale image** using NumPy. Each pixel intensity is a random value between 0 (black) and 255 (white).

### Step 1: Generate Random Image Matrix

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 10x10 random matrix with values between 0 and 255
img = np.random.randint(0, 256, (10, 10))
print(f"Image array:\n{img}")
````

This will produce a **10×10 array** where each number represents a pixel intensity.

---

### Step 2: Display the Image

```python
plt.imshow(img, cmap='gray')   # Display as grayscale
plt.title("Random Grayscale Image")
plt.axis('off')                # Hide axes for a cleaner look
plt.show()
```

---

### Explanation:

* **`np.random.randint(0, 256, (10,10))`** → Generates a 10×10 matrix with random integers from 0 to 255.
* **`plt.imshow(img, cmap='gray')`** → Converts the 2D array into a grayscale image.
* **Pixel intensity**:

  * 0 → Black
  * 255 → White
  * Values in between → Shades of gray
* **`plt.axis('off')`** → Hides the axes so the image looks clean.

Each time you run this code, a **new random pattern of grayscale pixels** will appear, giving you different random images.

---


`````python
img = np.random.randint(0, 256, (10,10))
print (f"Image array:\n{img}")
plt.imshow(img, cmap='gray')
plt.title("Random Grayscale Image")
plt.axis('off')  # Hide axes
plt.show()
`````

# Random Image Visualization (Default Colormap)

We generate a **10×10 random image** using NumPy:

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 10x10 random matrix with values between 0 and 255
img = np.random.randint(0, 256, (10,10))
print(f"Image array:\n{img}")

plt.imshow(img)           # Default colormap (not grayscale)

````

---

### Explanation:

1. **`np.random.randint(0, 256, (10,10))`**

   * Creates a 10×10 matrix of random pixel intensities (0–255).

2. **`plt.imshow(img)`**

   * By default, Matplotlib applies a **color map** (`viridis`) if `cmap` is not specified.
   * That means your "grayscale" numbers will now appear in **color** instead of shades of gray.

3. **`plt.axis('off')`**

   * Hides axes for a cleaner display.

---

💡 **Tip:**

* If you want a **true grayscale image**, use `cmap='gray'`:

```python
plt.imshow(img, cmap='gray')
```

* Without `cmap='gray'`, the image will be **pseudo-colored** based on Matplotlib’s default colormap.
