## **1. Import NumPy**

```python
import numpy as np
```

* NumPy is used for numerical operations and working with vectors/matrices.

---

## **2. Define Two Vectors**

```python
a = np.array([4, -1.5, 6])
b = np.array([3, 8, -4])
```

* `a` and `b` are **1D arrays**, representing vectors in 3-dimensional space.
* Vector `a` → `[4, -1.5, 6]`
* Vector `b` → `[3, 8, -4]`

---

## **3. Compute Dot Product**

```python
dot_product = np.dot(a, b)
```

* `np.dot(a, b)` calculates the **dot product** of vectors `a` and `b`.

**Mathematically**, for vectors ( \mathbf{a} = [a_1, a_2, a_3] ) and ( \mathbf{b} = [b_1, b_2, b_3] ):

[
\mathbf{a} \cdot \mathbf{b} = a_1b_1 + a_2b_2 + a_3b_3
]

* In your example:

[
4 \cdot 3 + (-1.5) \cdot 8 + 6 \cdot (-4) = 12 - 12 - 24 = -24
]

---

## **4. Print the Result**

```python
print(f"Dot product of {a} and {b} is: {dot_product}")
```

Output:

```
Dot product of [ 4.  -1.5  6. ] and [ 3  8 -4 ] is: -24.0
```

---

### **Summary**

| Operation     | Explanation                                                |
| ------------- | ---------------------------------------------------------- |
| `np.array`    | Create vectors                                             |
| `np.dot(a,b)` | Computes the sum of the products of corresponding elements |
| `print`       | Shows the result                                           |

---

````python
import numpy as np
a = np.array([4,-1.5,6])
b = np.array([3,8,-4])

dot_product = np.dot(a,b)
print(f"Dot product of {a} and {b} is: {dot_product}")
````

## **1. Create Random RGB Images**

```python
img_A = np.random.randint(0, 256, (50,50,3), dtype=np.uint8)
img_B = np.random.randint(0, 256, (50,50,3), dtype=np.uint8)
img_C = np.random.randint(0, 256, (50,50,3), dtype=np.uint8)
```

* Each image is a **50×50 RGB image**.
* `(50,50,3)` → 50 rows, 50 columns, 3 color channels (Red, Green, Blue).
* `dtype=np.uint8` → pixel values are integers from 0 to 255.

---

## **2. Flatten the Images**

```python
flattened_img_A = img_A.flatten()
flattened_img_B = img_B.flatten()
flattened_img_C = img_C.flatten()
```

* `.flatten()` converts each **3D image array** `(50,50,3)` → **1D array** of size `50*50*3 = 7500`.
* This allows us to treat the image as a **long vector**.

---

## **3. Compute Dot Product Similarity**

```python
similarity_AB = np.dot(flattened_img_A, flattened_img_B)
similarity_AC = np.dot(flattened_img_A, flattened_img_C)
```

* `np.dot()` calculates the **dot product** between flattened vectors.
* This acts as a **similarity measure**:

  * Higher dot product → more similar images in terms of pixel intensity alignment.
  * Lower dot product → less similar.

**Mathematical idea**:

[
\text{similarity} = \sum_{i=1}^{7500} A_i \cdot B_i
]

---

## **4. Compare Similarities**

```python
print(f"Similarity between Image A and Image B: {similarity_AB}")
print(f"Similarity between Image A and Image C: {similarity_AC}")

if similarity_AB > similarity_AC:
    print("Image A is more similar to Image B than to Image C.")
else :
    print("Image A is more similar to Image C than to Image B.")
```

* Compares `similarity_AB` and `similarity_AC`.
* Prints which image is closer to `Image A`.

---

###  **Summary**

| Step          | Operation                          | Purpose                            |
| ------------- | ---------------------------------- | ---------------------------------- |
| Create images | `np.random.randint`                | Generate 50×50 RGB images          |
| Flatten       | `.flatten()`                       | Convert 3D image → 1D vector       |
| Dot product   | `np.dot()`                         | Measure similarity between images  |
| Compare       | `if similarity_AB > similarity_AC` | Decide which image is more similar |

---

````python
img_A = np.random.randint(0, 256, (50,50,3), dtype=np.uint8)
img_B = np.random.randint(0, 256, (50,50,3), dtype=np.uint8)
img_C = np.random.randint(0, 256, (50,50,3), dtype=np.uint8)

flattened_img_A = img_A.flatten()
flattened_img_B = img_B.flatten()
flattened_img_C = img_C.flatten()

similarity_AB = np.dot(flattened_img_A, flattened_img_B)
similarity_AC = np.dot(flattened_img_A, flattened_img_C)

print(f"Similarity between Image A and Image B: {similarity_AB}")
print(f"Similarity between Image A and Image C: {similarity_AC}")

if similarity_AB > similarity_AC:
    print("Image A is more similar to Image B than to Image C.")
else :
    print("Image A is more similar to Image C than to Image B.")
````
